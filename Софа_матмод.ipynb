```python
import numpy as np
import math

def find_max_superdiagonal_element(A):
    """Находит индексы (i, j) максимального элемента над главной диагональю."""
    n = A.shape[0]
    max_val = -1
    i_max, j_max = 0, 1
    for i in range(n):
        for j in range(i + 1, n):
            val = abs(A[i, j])
            if val > max_val:
                max_val = val
                i_max, j_max = i, j
    return i_max, j_max, max_val

def compute_rotation_angle(A, i, j):
    """Вычисляет угол поворота φ для обнуления A[i,j]."""
    if A[i, i] == A[j, j]:
        fi = math.atan(1.5) / 2
    else:
        res = A[i, j] / (A[i, i] - A[j, j])
        fi = math.atan(res) / 2
    return fi

def create_rotation_matrix(n, i, j, fi):
    """Создаёт матрицу вращения U размера n×n для индексов i, j и угла fi."""
    U = np.eye(n, dtype=float)
    cos_fi = math.cos(fi)
    sin_fi = math.sin(fi)
    U[i, i] = cos_fi
    U[j, j] = cos_fi
    U[i, j] = -sin_fi
    U[j, i] = sin_fi
    return U

def compute_off_diagonal_norm(A):
    """Вычисляет евклидову норму внедиагональных элементов (над диагональю)."""
    n = A.shape[0]
    off_diag_elements = []
    for i in range(n):
        for j in range(i + 1, n):
            off_diag_elements.append(A[i, j])
    return math.sqrt(sum(x**2 for x in off_diag_elements))

def jacobi_eigenvalue_method(A, tolerance=0.1, max_iter=100):
    """Метод Якоби для нахождения собственных значений симметричной матрицы."""
    A = A.copy().astype(float)
    n = A.shape[0]
    count = 0

    while True:
        # Найти максимальный внедиагональный элемент
        i, j, max_val = find_max_superdiagonal_element(A)
        e = compute_off_diagonal_norm(A)

        if e <= tolerance or count >= max_iter:
            break

        # Вычислить угол вращения
        fi = compute_rotation_angle(A, i, j)

        # Построить матрицу вращения
        U = create_rotation_matrix(n, i, j, fi)

        # Обновить матрицу: A = U^T @ A @ U
        A = U.T @ A @ U

        count += 1
        print(f"Iteration {count}: off-diagonal norm = {e:.6f}")
        # print("Current A:\n", A)

    return A, count, e

# --- Основной код ---
if __name__ == "__main__":
    A = np.array([[-8, 9, 6],
                  [9, 9, 1],
                  [6, 1, 8]], dtype=float)

    print("Initial matrix A:\n", A)
    final_A, iterations, final_error = jacobi_eigenvalue_method(A, tolerance=0.1)

    print("\nFinal diagonalized matrix:\n", final_A)
    print(f"\nConverged in {iterations} iterations with off-diagonal norm = {final_error:.6f}")
    print("Approximate eigenvalues (diagonal elements):", np.diag(final_A))
```

```python
def f(x):
  return 1/(1+np.exp(-x))

x = np.linspace(-3, 3, 15)
np.random.seed(42)
e = np.random.normal(0, 0.1, 15)
y = f(x) + e

def get_polinoms (x_p, y_p, degree):
  X = np.vander(x_p, N=degree+1, increasing=True)
  coeffs, residuals, rank, s = np.linalg.lstsq(X, y_p, rcond=None)
  return coeffs

coeffs_p1 = get_polinoms(x, y, degree=1)
coeffs_p2 = get_polinoms(x, y, degree=2)
coeffs_p3 = get_polinoms(x, y, degree=3)

# print("P1 коэффициенты:", coeffs_p1)
# print("P2 коэффициенты:", coeffs_p2)
# print("P3 коэффициенты:", coeffs_p3)

def calc_poly(x, coeffs):
    return sum(c * (x ** i) for i, c in enumerate(coeffs))

errors_p1 = y - calc_poly(x, coeffs_p1)
errors_p2 = y - calc_poly(x, coeffs_p2)
errors_p3 = y - calc_poly(x, coeffs_p3)

# print("\nОшибки P1:", errors_p1)
# print("Ошибки P2:", errors_p2)
# print("Ошибки P3:", errors_p3)

plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = 1/(1+e^(-x))')
plt.plot(x, calc_poly(x, coeffs_p1), 'g-', label='P1(x)')
plt.plot(x, calc_poly(x, coeffs_p2), 'm-', label='P2(x)')
plt.plot(x, calc_poly(x, coeffs_p3), 'c-', label='P3(x)')
plt.scatter(x, y, color='red', label='Наблюдения с шумом')
plt.title('Исходная функция и набор точек')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)


plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
plt.scatter(x, errors_p1, color='green', label='Ошибки P1', s=60)
plt.scatter(x, errors_p2, color='magenta', label='Ошибки P2', s=60)
plt.scatter(x, errors_p3, color='cyan', label='Ошибки P3', s=60)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.xlabel('x')
plt.ylabel('Ошибка')
plt.title('График ошибок для полиномов P1, P2, P3')
plt.legend()
plt.grid(True)
plt.show()

polynomials = {
    'P1': errors_p1,
    'P2': errors_p2,
    'P3': errors_p3
}

print("Статистика ошибок для каждого полинома:\n")
print(f"{'Полином':<6} {'Среднее':<12} {'Стд. откл.':<12} {'Дисперсия':<12}")
print("-" * 50)

for name, errors in polynomials.items():
    mean_err = np.mean(errors)
    std_err = np.std(errors, ddof=1)
    var_err = np.var(errors, ddof=1)

    print(f"{name:<6} {mean_err:<12.6f} {std_err:<12.6f} {var_err:<12.6f}")

plt.figure(figsize=(12, 4))

for i, (name, errors) in enumerate(polynomials.items(), 1):
    plt.subplot(1, 3, i)
    plt.hist(errors, bins=6, color=['green','magenta','cyan'][i-1], alpha=0.7, edgecolor='black')
    plt.axvline(np.mean(errors), color='red', linestyle='--', label=f'Среднее = {np.mean(errors):.3f}')
    plt.title(f'Распределение ошибок: {name}')
    plt.xlabel('Ошибка')
    plt.ylabel('Частота')
    plt.grid(True, alpha=0.3)
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
def f(x):
  return np.exp(2*x) + 3*x - 4

x = np.linspace(0, 1, 100)
y = f(x)

plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
plt.plot(x, f(x), 'b-', linewidth=2, label='f(x) = e^2x + 3x - 4')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

def f1(x):
  return np.exp(2*x)

x_1 = np.linspace(0, 1, 100)
y_1 = f1(x)

def f2(x):
  return 4 - 3*x

x_2 = np.linspace(0, 1, 100)
y_2 = f2(x)

plt.plot(x_1, f1(x), 'r-', linewidth=2, label='f(x) = e^2x')
plt.plot(x_2, f2(x), 'g-', linewidth=2, label='f(x) = 4 - 3*x')
plt.show()

a, b = 0.0, 1.0
e = 10**-3
count = 0

print("Итерации:")
print(f"{'№':<3} {'a':<12} {'b':<12} {'c':<12} {'f(c)':<12} {'|f(b)-f(a)|':<12}")

while True:
    count += 1
    c = (a + b) / 2.0

    # width = abs(b - a)
    f_diff = abs(f(b) - f(a))

    print(f"{count:<3} {a:<12.6f} {b:<12.6f} {c:<12.6f} {f(c):<12.6f} {f_diff:<12.6f}")
    if f_diff < e:
        break
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c

root = (a + b) / 2
print(f"   Найден корень: x ≈ {root:.6f}")
print(f"   f(x) = {f(root):.2e}")
print(f"   Число итераций: {count}")
print(f"   Ширина отрезка: ")
print(f"   |f(b) - f(a)| = {abs(f(b) - f(a)):.2e}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(2 * x) + 3 * x - 4

def f_prime(x):
    return 2 * np.exp(2 * x) + 3

x_plot = np.linspace(0, 1, 200)
plt.figure(figsize=(8, 5))
plt.plot(x_plot, f(x_plot), 'b-', linewidth=2, label=r'$f(x) = e^{2x} + 3x - 4$')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0.5, color='orange', linestyle=':', label=r'$x_0 = 0.5$')
plt.scatter([0.5], [f(0.5)], color='orange', zorder=5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Метод Ньютона: выбор начального приближения')
plt.legend()
plt.grid(True)
plt.show()

x = 0.5
eps = 10**-3
max_iter = 20
count = 0

print("Метод Ньютона:")
print(f"{'k':<2} {'x_k':<15} {'f(x_k)':<15} {'f\'(x_k)':<15} {'|Δx|':<10}")
print("-" * 65)

while count < max_iter:
    x_new = x - f(x) / f_prime(x)
    dx = abs(x_new - x)

    print(f"{count:<2} {x:<15.10f} {f(x):<15.6e} {f_prime(x):<15.6f} {dx:<10.2e}")

    if dx < eps and abs(f(x)) < eps:
        x = x_new
        count += 1
        print(f"{count:<2} {x:<15.10f} {f(x):<15.6e} {f_prime(x):<15.6f} {'—':<10}")
        break

    x = x_new
    count += 1

print(f"   Корень: x = {x:.10f}")
print(f"   f(x) = {f(x):.2e}")
print(f"   Итераций: {count}")
print(f"   Проверка: f(x) * f'(x) = {f(x) * f_prime(x):.2e} > 0  {'Да' if f(x)*f_prime(x) > 0 else 'Нет'}")
```

```python
def gradF(rx, ry):
  res = np.zeros(2)
  res[0] = r_dfdx.evalf(20, subs={x:rx, y:ry})
  res[1] = r_dfdy.evalf(20, subs={x:rx, y:ry})
  return res
gradF(1, 2)
```

```python
# Гесиан матрца
def gesF(rx, ry):
  res = np.zeros((2, 2))
  res[0][0] = r_ddf_dxdx.evalf(20, subs={x:rx, y:ry})
  res[0][1] = r_ddf_dxdy.evalf(20, subs={x:rx, y:ry})
  res[1][0] = r_ddf_dydx.evalf(20, subs={x:rx, y:ry})
  res[1][1] = r_ddf_dydy.evalf(20, subs={x:rx, y:ry})
  return res
gesF(1, 2)
```
